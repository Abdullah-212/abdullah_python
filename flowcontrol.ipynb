```python
class standardaccount:
    
    __default_delevery = 7

    welcomemessage = 'welcome to aatomz account'


    def __init__(self,name,mobile,email):
        self.name = name
        self.mobile = mobile
        self.email = email
        print(self.welcomemessage)

    def shopping(self,item):
        print(f'{self.name} bought {item}')
    
    def delevery(self):
        print(f'your order will be delevered in {self.__default_delevery} days')

    def deactivateaccount(self):
        print(f'{self.name} account has been deactivated successfully')
```

```python
class premiumaccount(standardaccount):
    welcomemessage = 'welcome to aatomz **prime** account'

    def delevery(self):
        premium_delevery = 3
        print(f'your order will be delevered in {premium_delevery} days')

    def prime_video(self):
        print('prime video membership enabled')

    def cancel_membership(self):
        print(f'{self.name} has successfully cancelled the membership')
```
